```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("book ratings dataset/Books Data with Category Language and Summary/Preprocessed_data.csv")
```

```python
data
```

```python
data.drop(columns=["Unnamed: 0"], inplace=True)
data.head()
```

## Top 0.1% Most Popular Books in terms of Times Rated

```python
num_of_ratings = data.groupby("book_title").count()["rating"].reset_index().sort_values(by="rating", ascending=False)
```

```python
stats_num_ratings = num_of_ratings["rating"].describe(percentiles=[.999])
stats_num_ratings
```

```python
threshold_most_rated = stats_num_ratings["99.9%"]
most_rated_books = num_of_ratings[num_of_ratings["rating"]>=threshold_most_rated]
most_rated_books
```

## Top 0.1% Most Popular Books in terms of Sum of All Ratings

```python
sum_ratings = data.groupby("book_title")["rating"].sum().reset_index().sort_values(by="rating", ascending=False)
stats_sum = sum_ratings['rating'].describe(percentiles=[.999])
stats_sum
```

```python
threshold_sum = stats_sum["99.9%"]
top_books_sum = sum_ratings[sum_ratings["rating"] >= threshold_sum]
top_books_sum
```

## Top 0.1% Most Times Rated, Books Ranking in terms of Average Ratings 

```python
data_most_rated_only = data[data["book_title"].isin(most_rated_books["book_title"])].reset_index()
```

```python
top_books_avg = data_most_rated_only.groupby("book_title")["rating"].mean().reset_index().sort_values(by="rating", ascending=False)
top_books_avg
```
