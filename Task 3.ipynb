## Matrix of books against users

```python
matrix = data_modified.pivot_table(index="user_id", columns="book_title", values="rating")
matrix
```

```python
def subtract_row_mean(row):
    row_mean = np.nanmean(row)  # Calculate the mean excluding NaN values
    return row.apply(lambda x: x - row_mean if not np.isnan(x) else x)

# Apply the function to each row of the DataFrame
matrix_normalized = matrix.apply(subtract_row_mean, axis=1)



# Display the resulting DataFrame
matrix_normalized
```

```python
matrix_normalized.fillna(0, inplace=True)
```

```python
matrix_normalized
```

# Making user clusters based on user-user collaborative filtering using common book ratings

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_similarity = cosine_similarity(matrix_normalized)
```

```python
user_similarity
```

```python
user_similarity.shape
```

```python
def get_similar_users(user_id):
    index = np.where(matrix.index==user_id)[0][0]
    neareset_neighbours = sorted(list(enumerate(user_similarity[index])), key=lambda x:x[1], reverse=True)[1:]
    return neareset_neighbours
```

```python
similar_user_indices_with_similarity = get_similar_users(254)
similar_user_indices = [x[0] for x in similar_user_indices_with_similarity]
similar_user_indices
```

```python
def get_rating_prediction_for_nulls(user_id):
    if user_id in matrix.index:
        
        #find 3 most similar users who rated the unread books of this and predict a rating for all of them. Recommend the top 10 predicted rating books 
        
        user_index = np.where(matrix.index==user_id)[0][0]
        
        prediction_list = []
        for i in range(len(matrix.columns)):
            if matrix.isnull().iloc[user_index][i]:
                k=3# number of top similar users who have rated the book
                # print(f"k = {k}")
                count =0
                # print(f"count = {count}")
                sum_sim_mul_rat = 0
                
                sum_sim = 0
                predicted_rate = 0
                similar_user_indices_with_similarity = get_similar_users(user_id)
                
                for similar_user_index, similarity in similar_user_indices_with_similarity:
                    # print(f"similar user index is {similar_user_index}")
                    if(count==k):
                        break
                    else:
                        if matrix.isnull().iloc[similar_user_index][i]:
                            continue
                    
                        if (matrix.iloc[similar_user_index][i]==0 or similarity==0):
                            continue
                        
                        count+=1
                        sum_sim_mul_rat+= similarity*matrix.iloc[similar_user_index][i]
                        sum_sim += similarity
                        # print(f"rating of user_id {matrix.index[similar_user_index]} for book{book_index}: \"{matrix.columns[book_index]}\" is {matrix.iloc[similar_user_index][book_index]} \nsimilarity is {similarity} at count {count}")
                if sum_sim!=0:
                    predicted_rate = sum_sim_mul_rat/sum_sim
                    prediction_list.append((matrix.columns[i], predicted_rate))
                # print(f"for user:{user} for book:{matrix.columns[i]} predicted rate is {predicted_rate}")
        print(prediction_list)
        # return prediction_list
        return pd.DataFrame(prediction_list[:10], columns=["book_title", "predicted_rating"]).sort_values(by="predicted_rating", ascending=False)
    
    else:
        
        # Recommend top 10 popular books
        
        if user_id in data["user_id"].values:
            num_of_ratings = data.groupby("book_title").count()["rating"].reset_index().sort_values(by="rating", ascending=False)
            stats_num_ratings = num_of_ratings["rating"].describe(percentiles=[.999])
            threshold_most_rated = stats_num_ratings["99.9%"]
            most_rated_books = num_of_ratings[num_of_ratings["rating"]>=threshold_most_rated]
            
            data_most_rated_only = data[data["book_title"].isin(most_rated_books["book_title"])].reset_index()
            top_books_avg = data_most_rated_only.groupby("book_title")["rating"].mean().reset_index().sort_values(by="rating", ascending=False)[:10]
            return top_books_avg
            
            
        else:
            return "Can not make prediction for this user"
```
