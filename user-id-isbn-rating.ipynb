## Matrix of books against users

```python
matrix = data_modified.pivot_table(index="user_id", columns="isbn", values="rating")
matrix
```

```python
def subtract_row_mean(row):
    row_mean = np.nanmean(row)  # Calculate the mean excluding NaN values
    return row.apply(lambda x: x - row_mean if not np.isnan(x) else x)

# Apply the function to each row of the DataFrame
matrix_normalized = matrix.apply(subtract_row_mean, axis=1)



# Display the resulting DataFrame
matrix_normalized
```

```python
matrix_normalized.fillna(0, inplace=True)
```

```python
matrix_normalized
```

# Making user clusters based on user-user collaborative filtering using common book ratings

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_similarity = cosine_similarity(matrix_normalized)
```

```python
user_similarity
```

```python
user_similarity.shape
```

```python
def get_similar_users(user_id, suggestion_num):
    index = np.where(matrix.index==user_id)[0][0]
    neareset_neighbours = sorted(list(enumerate(user_similarity[index])), key=lambda x:x[1], reverse=True)[1:] # suggestion_num+1
    return neareset_neighbours
```

```python
similar_user_indices_with_similarity = get_similar_users(254, 0) # 0 in the parameter does not mean anything, not getting used
similar_user_indices = [x[0] for x in similar_user_indices_with_similarity]
similar_user_indices
```

```python
def get_rating_prediction(user_id, isbn):
    if user_id in matrix.index and isbn in matrix.columns:
        user_index = np.where(matrix.index==user_id)[0][0]
        k =3   # number of top similar users who have rated the book
        # print(f"k = {k}")
        count =0
        # print(f"count = {count}")
        sum_sim_mul_rat = 0
        
        sum_sim = 0
        predicted_rate = 0
        similar_user_indices_with_similarity = get_similar_users(user_id, 0) # 0 in the parameter does not mean anything, not getting used
        
        
        book_index = np.where(matrix.columns==isbn)[0][0]
        
        for similar_user_index,similarity in similar_user_indices_with_similarity:
            # print(f"similar user index is {similar_user_index}")
            if(count==k):
                break
            else:
                if matrix.isnull().iloc[similar_user_index][book_index]:
                    continue
                if (matrix.iloc[similar_user_index][book_index]==0 or similarity==0):
                    continue
                
                count+=1
                sum_sim_mul_rat+= similarity*matrix.iloc[similar_user_index][book_index]
                sum_sim += similarity
                # print(f"rating of user_id {matrix.index[similar_user_index]} for book{book_index}: \"{matrix.columns[book_index]}\" is {matrix.iloc[similar_user_index][book_index]} \nsimilarity is {similarity} at count {count}")
        if sum_sim==0:
            return "Not possible to rate due to insufficient data"
        else:
            predicted_rate = sum_sim_mul_rat/sum_sim
            return predicted_rate
        
    else:
        return "User or Book not available for prediction"

```
